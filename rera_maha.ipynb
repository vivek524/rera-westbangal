```python
from selenium import webdriver
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import os
import openpyxl
from openpyxl import load_workbook
url = "https://maharerait.mahaonline.gov.in/searchlist/search"
from selenium import webdriver
def driver_call(url):
        global driver
        #System.setProperty("webdriver.chrome.driver","/usr/bin/chromedriver");
        driver = webdriver.Chrome(executable_path="/home/anm/Downloads/chromedriver_linux64 (2)/chromedriver")
        #driver= webdriver.Chrome(executable_path="/home/anm/Downloads/chromedriver_linux64/chromedriver")
        driver.get(url)
        time.sleep(8)
def get_all_tables(soup):
    """Extracts and returns all tables in a soup object"""
    return soup.find_all("table")
def get_all_tables(soup):
    """Extracts and returns all tables in a soup object"""
    return soup.find_all("table")
def get_table_headers(table):
    """Given a table soup, returns all the headers"""
    headers = []
    for th in table.find("tr").find_all("th"):
        headers.append(th.text.strip())
    return headers
def get_table_rows(table):
    """Given a table, returns all its rows"""
    rows = []
    for tr in table.find_all("tr")[1:]:
        cells = []
        # grab all td tags in this table row
        tds = tr.find_all("td")
        if len(tds) == 0:
            # if no td tags, search for th tags
            # can be found especially in wikipedia tables below the table
            ths = tr.find_all("th")
            for th in ths:
                cells.append(th.text.strip())
        else:
            # use regular td tags
            for td in tds:
                cells.append(td.text.strip())
        rows.append(cells)
    return rows
def save_as_csv(table_name, headers, rows, i, writer):
    #pd.DataFrame(rows, columns=headers).to_csv(f"{table_name}.csv")
    # Create a Pandas Excel writer using XlsxWriter as the engine.
    #writer = pd.ExcelWriter('pandas_multiple.xlsx', engine='xlsxwriter')
   



    # Write each dataframe to a different worksheet.
    pd.DataFrame(rows, columns=headers).to_excel(writer, sheet_name=f"Sheet-{i}")
    writer.save()
    writer.close()
```

```python
i=0
driver_call(url)
Project_approved=[]
elementID=driver.find_element_by_xpath('//*[@id="TheForm"]/div[1]/div[3]/div/div/div[2]/div[2]')
elementID.click()
elementID=driver.find_element_by_xpath('//*[@id="Promoter"]')
elementID.click()
elementID=driver.find_element_by_xpath('//*[@id="btnAdvance"]')
elementID.click()
elementID=driver.find_element_by_xpath('//*[@id="State"]/option[4]')
elementID.click()
time.sleep(5)
elementID=driver.find_element_by_xpath('//*[@id="District"]/option[20]')
elementID.click()
elementID=driver.find_element_by_xpath('//*[@id="btnSearch"]')
elementID.click()
elementID=driver.find_element_by_xpath('//*[@id="gridview"]/div[1]/div/table/tbody/tr[1]/td[5]/b/a')
elementID.click()
driver.switch_to.window(driver.window_handles[-1])

Project_approved.append(driver.page_source)

soup = BeautifulSoup(driver.page_source, 'lxml')

tables = get_all_tables(soup)
print(f"[+] Found a total of {len(tables)} tables.")
TARGET_DIR = f'/home/anm/'

                    # iterate over all tables
filepath = "{}Project1_{}.xlsx".format(TARGET_DIR,i+1)
wb = openpyxl.Workbook()
wb.save(filepath)
print(filepath)

print("Created excel file Project1_{}.xlsx".format(i+1))
for num_tables, table in enumerate(tables, start=1):
    # get the table headers
    headers = get_table_headers(table)
    # get all the rows of the table
    rows = get_table_rows(table)
    path = "{}Project1_{}.xlsx".format(TARGET_DIR,i+1)
    book = load_workbook(path)
    writer = pd.ExcelWriter(path, engine = 'openpyxl')
    writer.book = book

    # save table as sheet
    table_name = f"table  -{num_tables}"
    print(f"[+] Saving {table_name}")
    save_as_csv(table_name, headers, rows, num_tables, writer)
```
